```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np



midwest = pd.read_csv("C:\\Users\\Rumana and Rishitha\\Downloads\\midwest (1).txt")

# Filter the dataset to include only rows with 'poptotal' less than 80000
midwest = midwest[midwest['poptotal'] < 80000]

# Create a figure and axis
fig, ax = plt.subplots()

# Group the data by 'state'
groups = midwest.groupby('state')

# Define a list of colors
colors = [(29, 150, 181), (173, 198, 237), (255, 126, 13), (255, 186, 121),
          (45, 161, 45), (151, 224, 139), (215, 35, 44), (255, 155, 151),
          (149, 104, 187), (198, 175, 214), (142, 85, 76), (191, 151, 146),
          (228, 117, 195), (246, 185, 212), (125, 126, 125), (195, 197, 198),
          (186, 184, 36), (217, 216, 142), (25, 197, 205), (157, 217, 227)]

# Normalize the colors to values between 0 and 1
colors = [(r / 255., g / 255., b / 255.) for r, g, b in colors]

# Select every other color
colors = colors[::2]

# Plot each group
for i, (name, group) in enumerate(groups):
    group.plot(
        kind='scatter',
        x='area',
        y='poptotal',
        ylim=(0, 80000),
        xlim=(0., 0.1),
        s=10 + group['popdensity'] * 0.1,
        label=name,
        ax=ax,
        color=colors[i],
        edgecolor='w'
    )

# Create a legend for state colors
lgd = ax.legend(numpoints=1, loc=1, borderpad=1,
               frameon=True, framealpha=0.9, title="State")

# Set the marker sizes in the legend
for handle in lgd.legendHandles:
    handle.set_sizes([100.0])

# Create a legend for popdensity
pws = (pd.cut(midwest['popdensity'], bins=4, retbins=True)[1]).round(0)
for pw in pws:
    plt.scatter([], [], s=(pw ** 2) / 2e4, c="k", label=str(pw))

# Display the plot
plt.show()
```
